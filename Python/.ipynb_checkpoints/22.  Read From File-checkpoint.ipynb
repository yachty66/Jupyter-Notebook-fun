```python
#Exercise
x=[]
counter_darth = 0
counter_lea = 0
counter_luke = 0

with open('nameslist.txt', 'r') as open_file:
    line = open_file.readline()
    while line:
        x.append(line)
        if line == 'Darth\n':
            counter_darth += 1
        elif line == 'Luke\n':
            counter_luke += 1
        elif line == 'Lea\n':
            counter_lea += 1
        line = open_file.readline()
#print(x)


print("Darth: " + str(counter_darth) + ", Luke: " + str(counter_luke) + ", Lea: " + str(counter_lea))
```
