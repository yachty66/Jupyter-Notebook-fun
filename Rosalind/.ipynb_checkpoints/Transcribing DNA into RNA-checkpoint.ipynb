```python
def countRNANucleotides():
    #jedes T in einem uebergebenen string muss mit einem u ausgetauscht werden
    RNA = input()
    final = []
    for i in RNA:
        if(i == "T"):
            final.append("U")
        else:
            final.append(i)
    return "".join(final)
            
```

```python
print(countRNANucleotides())
```
