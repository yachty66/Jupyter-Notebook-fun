```python
from newspaper import Article
import nltk
from gtts import gTTS
import os
```

```python
article =  Article('https://openblog.ai/blog_article/Backpropagation/Backpropagation.html')
article.download()
article.parse()
```

```python
nltk.download("punkt")
article.nlp()
```

```python
mytext = article.text
language = "en"
myobj = gTTS(text=mytext, lang=language, slow=False)
```

```python
myobj.save("read_article2.mp3")
os.system("start read_article2.mp3")
```
